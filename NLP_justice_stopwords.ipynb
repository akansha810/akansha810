```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
nlp = spacy.load("en_core_web_sm")
##  preprocess function which takes a string then remove the stopwords and then returns another string
def preprocess(text):
  doc= nlp(text)

  no_stop_words =[token.text for token in doc if not token.is_stop and not token.is_punct]
  return " ".join(no_stop_words)
```

```python
justice["contents_new"]= justice["contents"].apply(preprocess)
```
